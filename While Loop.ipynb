```
n = int(input("Enter a number: "))
sum = 0
i = 1
while i<=n:
  sum = sum+i
  i = i+1
print(f"Sum = {sum}")
```

```
num = int(input("Enter a number: "))
sum = 0
temp = num
while temp > 0:
  digit = temp % 10
  sum += digit ** 3
  temp //= 10
if sum == num:
  print(f"{num} is an Armstrong Number!")
else:
  print(f"{num} is not an Armstrong Number!")
```
