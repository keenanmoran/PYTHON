```
try:
  num1, num2 = eval(input("What are the two numbers? "))
  result = num1/num2
  print(result)
except ZeroDivisonError:
    print("Divison by 0 is error")
except SyntaxError:
    print("The format is 1, 2")
except:
    print("Wrong input")
else:
    print("No exceptions")
finally:
    print("This will excecute no matter what")
```

```
try:
  number = int(input("Enter a number: "))
  print("The number entered is", number)
except ValueError as ex:
  print("Exception: ", ex)
```

```
valid = False
while not valid:
  try:
    n = int(input("Enter a number: "))
    while n % 2 == 0:
      print("bye")
    valid = True
  except ValueError:
    print("Invalid Input")
```
