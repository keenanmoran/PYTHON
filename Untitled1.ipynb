```python
def match_words(word):
  ctr = 0
  lst = []
  for i in word:
    if len(i)>1 and i[0]==i[-1]:
      ctr = ctr+1
      lst.append(i)
  return ctr
words = input("Input a sentence: ").split()
match_words(words)
```
