```python
import random
playing = True
print("Rock, Paper, Scissors")
print("="*40)
print(" ")
while playing:
  rps = ["rock","paper","scissors"]
  choice = random.choice(rps)
  player = input("Are you ready? When I say shoot, you will type in either rock, paper, or scissors. Rock, Paper, Scissors, SHOOT! ")
  if choice == player:
    print("Its a tie!")
  elif player == "rock":
    if choice == "scissors":
      print("You win!")
    else:
      print("You lose!")
  elif player == "scissors":
    if choice == "rock":
      print("You lose!")
    else:
      print("You win!")
  elif player == "paper":
    if choice == "rock":
      print("You win!")
    else:
      print("You lose")
  else:
    print("Invalid input:(")
```
