```python
weather = (1, 0, 0, 0, 1, 1, 1)
sunny = 0
rainy = 0
for i in weather:
  if i == 0:
    rainy = rainy+1
  elif i == 1:
    sunny = sunny+1
  else:
    print("Invalid data")
if sunny > rainy:
  print("Sunny weather")
elif rainy > sunny:
  print("Rainy weather")
else:
  print("Invalid data")
```
